```python
data["service_amount_net"] = data["service_amount_net"].str.replace(",", ".").astype("float")
```

```python
data["service_amount_net"]
```

```python
data.info()
```

```python
def clearTimeData(x):
    temp = []

    for item in x:
        i = item.find(".")
        temp.append(item[:i])

    return pd.Series(data=temp, name=x.name)
```

```python
data['service_date'] = pd.to_datetime(clearTimeData(data['service_date']), format="%Y-%m-%d %H:%M:%S")
```

```python
data
```

```python
data.info()
```

```python
round_to_day = data['service_date'].dt.to_period('M').dt.to_timestamp()
data["service_date_month"] = round_to_day
data
```

```python
round_to_day = data['service_date'].dt.to_period('D').dt.to_timestamp()
data["service_date_day"] = round_to_day
data
```

### **С ЭТОГО МОМЕНТА СТРОЯТСЯ ГРАФИКИ**

```python
sns.displot(data=data[data['service_amount_net'] > 20000], x="service_amount_net")
plt.title("Сумма выплаты > 20000")
```

```python
fig, ax = plt.subplots(2, 3, figsize=(23, 10))

plt.subplots_adjust(hspace = 0.3)

for i, threshold in enumerate(range(0, 30000, 5000)):
    sns.kdeplot(data=data[data['service_amount_net'] > threshold], x="service_amount_net", ax=ax[i//3][i%3])
    ax[i//3][i%3].set_title(f"Сумма выплат > {threshold}")
    ax[i//3][i%3].grid()

plt.show()
```

### Тут оперативка коллаба умерла $\downarrow$
