```python
import pandas as pd
data = pd.read_csv('CupIT_Sber_data.csv', sep=';')
def add_date_time_columns(df, date_column):
    # Преобразуем колонку в datetime, если она еще не в этом формате
    df[date_column] = pd.to_datetime(df[date_column])

    # Создаем новые колонки
    df['year'] = df[date_column].dt.year  # Год
    df['month'] = df[date_column].dt.month  # Месяц
    df['date'] = df[date_column].dt.date  # Дата (год, месяц, день)
    df['time'] = df[date_column].dt.time  # Время (часы, минуты, секунды)

    return df
data = add_date_time_columns(data, 'service_date')
data['service_amount_net'] = data['service_amount_net'].str.replace(',', '.', regex=False)
data.head
```

```python
#день недели
from datetime import datetime
import numpy as np
days_ru = {
    0: "ПН",
    1: "ВТ",
    2: "СР",
    3: "ЧТ",
    4: "ПТ",
    5: "СБ",
    6: "ВС"
}
data['weekday_name'] = np.vectorize(lambda x: days_ru[x.weekday()])(data['date'])
data.head()
```

```python
#выходные
#1-выходной, 0-нет

def wknd(x):
    if x == 'СБ' or x == 'ВС':
        return 1
    else:
        return 0
data['weekends'] = np.vectorize(wknd)(data['weekday_name'])
data.head()
```

```python
#праздники
#1-праздник, 0-нет
def hld(x):
    m = [
        '2022-01-01',
        '2022-01-02',
        '2022-01-03',
        '2022-01-04',
        '2022-01-05',
        '2022-01-06',
        '2022-01-07',
        '2022-01-08',
        '2022-02-23',
        '2022-03-08',
        '2022-05-01',
        '2022-05-09',
        '2022-06-12',
        '2022-11-04',
        '2022-05-03',
        '2022-05-10',
        '2022-03-07',
        '2023-01-01',
        '2023-01-02',
        '2023-01-03',
        '2023-01-04',
        '2023-01-05',
        '2023-01-06',
        '2023-01-07',
        '2023-01-08',
        '2023-02-23',
        '2023-03-08',
        '2023-05-01',
        '2023-05-09',
        '2023-06-12',
        '2023-11-04',
        '2023-02-24',
        '2023-05-08'
    ]
    if x in m:
        return 1
    else:
        return 0
data['holidays'] = np.vectorize(hld)(data['date'].astype(str))
```

```python
df_sc = pd.DataFrame(columns = ['code', 'mean', 'median', 'std'])
for sc in data['service_code'].value_counts().keys().astype(str):
    t = data[data['service_code'] == sc]['service_amount_net'].astype(float)
    df_sc.loc[len(df_sc)] = [sc, t.mean(), np.median(t), t.std()]
    break
df_sc.head()
```

```python
df_sn = pd.DataFrame(columns = ['code', 'mean', 'median', 'std'])
for sn in data['service_name'].value_counts().keys().astype(str):
    t = data[data['service_name'] == sn]['service_amount_net'].astype(float)
    df_sn.loc[len(df_sn)] = [sn, t.mean(), np.median(t), t.std()]
df_sn.head()
```
